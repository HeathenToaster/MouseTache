```python
import subprocess

# Source and destination directories
source_base = "/NAS02/KenzaData/"
destination_base = "/LocalData/ForagingMice/4TowerTaskMethodPaper/KenzaData/"

# List of folders to transfer
folders_to_transfer = ['MOUKA267', 'MOUKA269', 'MOUKA271', 'MOUKA274', 'MOUKA276', 
                       'MOUKA277', 'MOUKA279', 'MOUKA280', 'MOUKA281', 'MOUKA282', 
                       'MOUKA283', 'MOUKA284', 'MOUKA285', 'MOUKA299', 'MOUKA300', 
                       'MOUKA303', 'MOUKA304', 'MOUKA305']

# Function to run rsync command
def run_rsync(source, destination):
    try:
        subprocess.run(['rsync', '-avz', source, destination], check=True)
        print(f"Successfully transferred {source} to {destination}")
    except subprocess.CalledProcessError as e:
        print(f"Error transferring {source} to {destination}: {e}")

# Loop through each folder and use rsync to transfer it
for folder in folders_to_transfer:
    source_folder = source_base + folder + '/'
    destination_folder = destination_base + folder + '/'
    run_rsync(source_folder, destination_folder)
```

```python
import subprocess

# Source and destination directories
source_base = "/NAS02/KenzaData/"
destination_base = "/media/david/Lexar/KenzaData/"

# List of folders to transfer
folders_to_transfer = ['MOUKA267', 'MOUKA269', 'MOUKA271', 'MOUKA274', 'MOUKA276', 
                       'MOUKA277', 'MOUKA279', 'MOUKA280', 'MOUKA281', 'MOUKA282', 
                       'MOUKA283', 'MOUKA284', 'MOUKA285', 'MOUKA299', 'MOUKA300', 
                       'MOUKA303', 'MOUKA304', 'MOUKA305']

# Function to run rsync command, excluding .avi files
def run_rsync(source, destination):
    try:
        subprocess.run(['rsync', '-avz', '--exclude', '*.avi', source, destination], check=True)
        print(f"Successfully transferred {source} to {destination} (excluding .avi files)")
    except subprocess.CalledProcessError as e:
        print(f"Error transferring {source} to {destination}: {e}")

# Loop through each folder and use rsync to transfer it, excluding .avi files
for folder in folders_to_transfer:
    source_folder = source_base + folder + '/'
    destination_folder = destination_base + folder + '/'
    run_rsync(source_folder, destination_folder)
```
